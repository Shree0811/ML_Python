```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
%matplotlib inline
```

```python
# Importing Iris data set for further analysis
from sklearn.datasets import load_iris
```

```python
Data= load_iris()
print(Data)
```

```python
# Extracting independent features from data set
Data.data
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
Model=DecisionTreeClassifier()
```

```python
# Make train and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(Data.data,Data.target , test_size=0.20, 
                                                            shuffle=True, random_state=2)
```

```python
# Now we will Train our Decision tree
# Now first we will set aprroriate figure size for tree
Model.fit(X_train,y_train)
plt.figure(figsize=(15,10))
tree.plot_tree(Model,filled=True) # filled=True will add colours to the different block
```

```python
# Now we will make prediction based on created model
Y_Pr=Model.predict(X_test)
# Y_Pr are the predicted values of by Decision tree

Y_Tr=y_test
# Y_Tr are the true values
```

```python
# Now lets assess the performace of decision tree
# Let's calculate accuracy for the prediction
Accuracy=sklearn.metrics.accuracy_score(Y_Tr, Y_Pr)
print("Accuracy of the Decision tree for the Prediction is :",round(Accuracy,4)*100,"%")
```
